Problem Statement:

You are provided with a DataFrame named df containing transaction data:

CustomerID  	TransactionDate	Amount	 Region

101   2023-01-10	200	      East

102	           2023-01-20	300      North

103	           2023-02-15	150	    South

104            2023-02-20	450	    West

105	           2023-03-10	350	    East

101	           2023-04-05	300	    East

102	           2023-04-12	500	    North

Tasks:

Identify the top-spending customer each month:

Group by Month and find the CustomerID with the highest total Amount spent in each month.

Calculate the month-over-month spending difference for each region:

For each Region, compute the difference in total spending between consecutive months.

Trend Analysis:

Based on the month-over-month differences:

Assign "Increase" if spending went up, "Decrease" if it went down, and "Same" if no change occurred.

Resulting DataFrame:

Add a new column SpendingTrend in the grouped result indicating the spending trend for each region.


```python
import pandas as pd
```

```python
d1={"CustomerID":[101,102,103,104,105,101,102],
"TransactionDate":["2023-01-10","2023-01-20","2023-02-15","2023-02-20","2023-03-10","2023-04-05","2023-04-12"],
"Amount":[200,300,150,450,350,300,500],
"Region":["East","North","South","West","East","East","North"]}
df=pd.DataFrame(d1)
df["TransactionDate"]=pd.to_datetime(df["TransactionDate"])
```

```python
df["Month"]=df["TransactionDate"].dt.to_period('M')
s=df.groupby(["CustomerID","Month"])["Amount"].sum().reset_index()
sf=s.loc[s.groupby('Month')['Amount'].idxmax()]
sf
```

```python
sf=df.groupby(["Region","Month"])["Amount"].sum().reset_index()
sf["Trend"]=sf.groupby("Region")["Amount"].diff()
sf
```

```python
def trendanalysis(Trend):
    if Trend >0:
        return "Increase"
    elif Trend <0:
        return "Decrease"
    else:
        return "Same"
sf["SpendingTrend"]=sf["Trend"].apply(trendanalysis)
df["SpendingTrend"]=sf["SpendingTrend"]
df
```
