Problem: Monthly Spending Trend by Region.

You are tasked with analyzing the monthly spending trend for customers in different regions. Specifically, you need to:

1.Group the data by month and region.

2.Calculate the total spending (sum of Amount) for each region in each month.

3.Identify the region with the highest spending in each month.

4.Identify whether the spending in each region increased, decreased, or remained the same compared to the previous month.

```python
import pandas as pd
```

```python
data = {
    "CustomerID": [101, 102, 103, 104, 105, 101, 102, 104],
    "TransactionDate": [
        "2023-01-10", "2023-02-15", "2023-02-20", 
        "2023-03-10", "2023-03-25", "2023-04-01",
        "2023-04-05", "2023-04-12"
    ],
    "Amount": [250, 300, 150, 500, 350, 200, 450, 300],
    "Region": ["East", "North", "South", "West", "East", "East", "North", "West"]
}

df = pd.DataFrame(data)
df["TransactionDate"] = pd.to_datetime(df["TransactionDate"])
```

```python
df["Month"]=df["TransactionDate"].dt.month
df.sort_values(["Month","Region"])
```

```python
df.groupby(["Month","Region"])["Amount"].sum()
```

```python
monthly_region_spending = df.groupby(["Month", "Region"])["Amount"].sum().reset_index()
highest_region_per_month = monthly_region_spending.loc[monthly_region_spending.groupby("Month")["Amount"].idxmax()]
highest_region_per_month
```

```python
monthly_region_spending["Trend"]=(
    monthly_region_spending.groupby("Region")["Amount"].diff().apply(lambda x: "Increase"if x>0 else ("Decrease" if x<0 else "Same"))
)
monthly_region_spending
```
