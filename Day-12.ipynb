```python
import pandas as pd
```

```python
d1={"Name":["John","Alice","Bob","Emma"],
"Age":[20,22,12,21],
"Gender":["Male","Female","Male","Female"],
"Marks":[88,92,75,89],
"City":["New York","Los Angeles","Chicago","San Fransico"]}
vr=pd.DataFrame(d1)
vr
```

```python
import numpy as np
vr.loc[vr["Name"]=="Bob","Age"]=np.nan
```

```python
vr["Age"]=vr["Age"].fillna(0)
vr
```

```python
def mark(row):
    if row["City"]=="Los Angeles" and row["Marks"]<=80:
        return row["Marks"]+5
    else:
        return row["Marks"]
vr["Marks"]=vr.apply(mark,axis=1)
vr
```

```python
def sorte(row):
    if row["Age"]<18:
        return "Disclosure"
    else:
        return row["City"]
vr["City"]=vr.apply(sorte,axis=1)
vr
```

```python
vr.loc[(vr["Marks"]>85) & (vr["City"].isin(["Chicago","San Fransico"]))]
```

```python
vr.sort_values(by=["Marks","Age"],ascending=[False,True])
```

```python
def grading(Marks):
    if Marks>90:
        return "Distinction"
    elif Marks>50:
        return "Pass"
    else:
        return "Fail"
vr["Grade"]=vr["Marks"].apply(grading)
```

```python
np.random.seed(42)
vr["Joindate"]=pd.to_datetime(np.random.choice(pd.date_range("2023-1-1","2023-12-31")))
vr["Month"]=vr["Joindate"].dt.month
m=vr.groupby("Month")["Marks"].sum()
```

```python
vr.groupby("Gender")["City"].count()
```

```python
def f(row):
    if row["Marks"]>80:
        return row["Marks"]*1.10
    else:
        return row["Marks"]*1.05
vr["FinalScore"]=vr.apply(f,axis=1)
vr
        
```
