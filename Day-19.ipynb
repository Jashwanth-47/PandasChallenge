1.Identify High-Value Transactions:

Create a column to classify transactions as "High-Value" if the amount exceeds 400, and "Low-Value" otherwise.

2.Find the Most Frequent Transaction Month by Region:

Determine the most common month for transactions in each region.

3.List of Unique Customers by Quarter:

For each quarter, list all unique CustomerIDs that made transactions.

4.Average Spending Per Quarter for Each Customer:

Calculate the average transaction amount per quarter for each CustomerID.

5.Days Between Consecutive Transactions:

For each CustomerID with multiple transactions, calculate the number of days between their consecutive transactions.

```python
import pandas as pd
```

```python
data = {
    "CustomerID": [101, 102, 103, 104, 105, 101, 102, 104],
    "TransactionDate": [
        "2023-01-10", "2023-02-15", "2023-02-20", 
        "2023-03-10", "2023-03-25", "2023-04-01",
        "2023-04-05", "2023-04-12"
    ],
    "Amount": [250, 300, 150, 500, 350, 200, 450, 300],
    "Region": ["East", "North", "South", "West", "East", "East", "North", "West"]
}

df = pd.DataFrame(data)
df["TransactionDate"] = pd.to_datetime(df["TransactionDate"])
```

```python
def func(Amount):
    if Amount>400:
        return "High-Value"
    else:
        return "Low-Value"
df["Transaction"]=df["Amount"].apply(func)
df
```

```python
df["Month"]=df["TransactionDate"].dt.month
df.groupby("Region")["Month"].apply(lambda x: x.value_counts().idxmax())
```

groupby("Region")["Month"]: 

Groups the data by region and accesses the "Month" column.

apply(lambda x: x.value_counts().idxmax()):

x.value_counts() counts the occurrences of each month in the group.

.idxmax() retrieves the month with the highest frequency.

This will give the most frequent (mode) month for each region without explicitly using mode.

```python
df["Month"]=df["TransactionDate"].dt.month
def s(Month):
    if Month in [1,2,3]:
        return "Q1"
    elif Month in [4,5,6]:
        return "Q2"
    elif Month in [7,8,9]:
        return "Q3"
    else:
        return "Q4"
df["Quarter"]=df["Month"].apply(s)
df.groupby("Quarter")["CustomerID"].unique()
```

```python
df.groupby(["Quarter","CustomerID"])["Amount"].mean()
```

```python
# Sort the DataFrame by CustomerID and TransactionDate to ensure correct ordering
df_sorted = df.sort_values(by=["CustomerID", "TransactionDate"])

# Calculate the difference in days for consecutive transactions per CustomerID
df_sorted["DaysBetweenTransactions"] = df_sorted.groupby("CustomerID")["TransactionDate"].diff().dt.days

# Display the result
df_sorted
```

```python
df["MonthDiff"]=df.groupby("CustomerID")["Month"].diff()
df.sort_values(by="MonthDiff") #just testing difference by month
```
