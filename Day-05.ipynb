```python
#Problem 1: Selecting Data
#a) Print all rows where Marks >= 50.
#b) Print the Name and City of students who scored less than 40.

#Problem 2: Updating Data
#a) Set City of "Alice" to "Chicago".
#b) Replace all Marks greater than 90 with 100.

#Problem 3: Adding and Modifying Columns
#Add a new column called Scholarship:
#Students get "Yes" if Marks >= 85, otherwise "No".

#Problem 4: Sorting Data
#Sort the DataFrame by Marks in ascending order and display the result.'''
```

```python
import pandas as pd
```

```python
d1={"Name":["John","Alice","Bob","Emma"],
"Age":[20,22,12,21],
"Gender":["Male","Female","Male","Female"],
"Marks":[88,92,75,89],
"City":["New York","Los Angeles","Chicago","San Fransico"]}
vr=pd.DataFrame(d1)
```

```python
vr.query("Marks>=50")
```

```python
vr.query("Marks<40")[["Name","City"]]
```

```python
vr.loc[vr["Name"]=="Alice","City"]="Chicago"
```

```python
def replace(Marks):
    if Marks >=90:
        return 100
    else:
        return Marks
vr["Marks"]=vr["Marks"].apply(replace)
vr
```

```python
def scholarship(Marks):
    if Marks >=85:
        return "Yes"
    else:
        return "No"
vr["Scholarship"]=vr["Marks"].apply(scholarship)
vr
```

```python
vr.sort_values(by="Marks")
```
