Problem:

A bank is analyzing transaction data to understand regional spending trends and customer activity.

1.Top Customers by Spending (Month-wise):

For each Month, find the CustomerID with the highest transaction amount.
Create a new DataFrame containing columns: Month, CustomerID, and Amount.

2.Trend Analysis by Region:

Calculate the trend in spending (Increase, Decrease, or Same) for each Region over the Month.
Add a Trend column to store this information.

```python
import pandas as pd 
data = {
    "CustomerID": [101, 102, 103, 104, 105, 106, 107, 108],
    "TransactionDate": [
        "2023-01-15", "2023-01-20", "2023-02-12", "2023-03-10", 
        "2023-03-12", "2023-03-15", "2023-04-10", "2023-04-20"
    ],
    "Region": ["East", "North", "South", "East", "North", "South", "East", "North"],
    "Amount": [200, 300, 150, 400, 350, 180, 250, 100]
}
df = pd.DataFrame(data)
df["TransactionDate"] = pd.to_datetime(df["TransactionDate"])
```

```python
df["Month"]=df["TransactionDate"].dt.month
transaction=df.groupby(["CustomerID","Month"])["Amount"].sum().reset_index()
highest_transaction=transaction.loc[transaction.groupby("Month")["Amount"].idxmax()]
highest_transaction
```

```python
df2=df[['Month','CustomerID','Amount','Region']].copy()
df2
```

```python
spending=df.groupby(["Month","Region"])["Amount"].sum().reset_index()
spendingdifference=spending.groupby("Region")["Amount"].diff()
def trendan(spendingdifference):
    if spendingdifference>0:
        return "Increase"
    elif spendingdifference<0:
        return "Decrease"
    else:
        return "Same"
spending["Trend"] = spendingdifference.apply(trendan)

# Merge the trend information back to df2 based on Region and Month.
df2 = df2.merge(spending[["Month", "Region", "Trend"]], on=["Month", "Region"], how="left")
df2
```
