Problem: Analyzing Sales Trends

You have a dataset representing the sales data of customers over time. Your goal is to identify sales trends for each customer by finding out:

Monthly Average Spend per Customer: 

Calculate the monthly average amount spent by each customer (CustomerID).

Spending Trend:

Compare each customerâ€™s current month's spending with the previous month.

Create a new column called Trend:

"Increase" if spending increased.

"Decrease" if spending decreased.

"Same" if spending stayed the same.

```python
import pandas as pd
```

```python
data = {
    "CustomerID": [101, 102, 103, 104, 105, 101, 102, 104],
    "TransactionDate": [
        "2023-01-10", "2023-02-15", "2023-02-20", 
        "2023-03-10", "2023-03-25", "2023-04-01",
        "2023-04-05", "2023-04-12"
    ],
    "Amount": [250, 300, 150, 500, 350, 200, 450, 300],
    "Region": ["East", "North", "South", "West", "East", "East", "North", "West"]
}

df = pd.DataFrame(data)
df["TransactionDate"] = pd.to_datetime(df["TransactionDate"])
```

```python
df["Month"]=df["TransactionDate"].dt.month
df.groupby(["CustomerID","TransactionDate"])["Amount"].mean().reset_index()
```

```python
df.sort_values(by=["CustomerID", "TransactionDate"], inplace=True)
df["MonthPeriod"]=df["TransactionDate"].dt.to_period('M')
df["Trend"]=df.groupby("CustomerID")["Amount"].diff()
def s(Trend):
    if Trend>0:
        return "Increase"
    elif Trend<0:
        return "Decrease"
    elif Trend==0:
        return "Same"
    else:
        return  "N/A"
df["Trend"]=df["Trend"].apply(s)
df
```
