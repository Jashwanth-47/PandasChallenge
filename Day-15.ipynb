Problem 1: Updating and Cleaning Data
Replace missing values in the City column with "Unknown."
Assign a value of -1 to any missing Marks.
Capitalize all names in the Name column.

Problem 2: Conditional Calculations
Create a column Scholarship:
Assign "Eligible" if Marks > 85.
Assign "On Review" if 75 <= Marks <= 85.
Assign "Not Eligible" if Marks < 75.
Add a column Category:
Assign "Minor" if Age < 18.
Assign "Adult" otherwise.

Problem 3: Filtering and Analysis
Filter students where:
Marks > the 75th percentile.
City is "Chicago" or "San Fransico".
Sort the filtered DataFrame by Marks in ascending order.

Problem 4: Grouping and Aggregation
Find the average Marks for each Performance category (e.g., "Outstanding", "Above Average").
Determine the maximum Age for each Quarter.

Problem 5: Advanced Date Operations
Calculate the difference in days between JoinDate and today's date for each student. Save this in a column DaysSinceJoined.
Identify the oldest JoinDate for each City using a groupby operation.


```python
import pandas as pd
```

```python
d1={"Name":["John","Alice","Bob","Emma"],
"Age":[14,17,12,21],
"Gender":["Male","Female","Male","Female"],
"Marks":[88,92,75,89],
"City":["New York","Los Angeles","Chicago","San Fransico"]}
vr=pd.DataFrame(d1)
```

```python
import numpy as np
vr.loc[vr["Name"]=="Bob","City"]=np.nan
vr["City"]=vr["City"].fillna("Unknown")
vr
```

```python
vr.loc[vr["Name"]=="Alice","Marks"]=np.nan
vr["Marks"]=vr["Marks"].fillna(-1)
vr
```

```python
vr["Name"]=vr["Name"].str.upper()
```

```python
def d(Marks):
    if Marks >85:
        return "Eligible"
    elif Marks >75:
        return "On Review"
    else:
        return "Not Eligible"
vr["Scholarship"]=vr["Marks"].apply(d)
def s(Age):
    if Age<18:
        return "Minor"
    else:
        return "Adult"
vr["Category"]=vr["Age"].apply(s)
vr
```

```python
vr.loc[(vr["Marks"]>75)&(vr["City"].isin(["Chicago","San Fransico"]))]
```

```python
vr.sort_values(by="Marks")
```

```python
def r(row):
    if row["Gender"]=="Female" and row["Marks"]>90:
        return "Outstanding"
    elif row["Gender"]=="Male" and row["Marks"]>85:
        return "Above Average"
    else:
        return "Average"

vr["Performance"]=vr.apply(r,axis=1)
vr.groupby("Performance")["Marks"].mean()
```

```python
import numpy as np 
np.random.seed(42)
vr["JoinDate"]=pd.to_datetime(np.random.choice(pd.date_range("2023-1-1","2023-12-31"),len(vr)))
vr["JoinMonth"]=vr["JoinDate"].dt.month
def s(JoinMonth):
    if JoinMonth in [1, 2, 3]:
        return "Q1"
    elif JoinMonth in [4, 5, 6]:
        return "Q2"
    elif JoinMonth in [7, 8, 9]:
        return "Q3"
    else:
        return "Q4"
vr["Quarter"]=vr["JoinMonth"].apply(s)
vr.groupby("Quarter")["Age"].max()
```

```python
today=pd.Timestamp.now()
vr["Dayssincejoined"]=abs(vr["JoinDate"]-today)
vr
```
