Tasks

1.Add a new column Month extracted from the TransactionDate column.

2.Identify the highest transaction (max amount) for each Region within each Month.

3.Determine the trend of spending for each Region across months, similar to your earlier problems:

"Increase": If the amount increases compared to the previous month.

"Decrease": If the amount decreases.

"Same": If the amount remains unchanged.

"N/A": For the first entry or missing data.


```python
import pandas as pd
```

```python
data = {
    "CustomerID": [101, 102, 103, 104, 105, 106, 107, 108],
    "TransactionDate": [
        "2023-01-15", "2023-01-20", "2023-02-12", "2023-03-10", 
        "2023-03-12", "2023-03-15", "2023-04-10", "2023-04-20"
    ],
    "Region": ["East", "North", "South", "East", "North", "South", "East", "North"],
    "Amount": [200, 300, 150, 400, 350, 180, 250, 100]
}
df = pd.DataFrame(data)
df["TransactionDate"] = pd.to_datetime(df["TransactionDate"])
```

```python
df["Month"]=df["TransactionDate"].dt.month
df
```

```python
s=df.groupby(["Region","Month"])["Amount"].sum().reset_index()
highest_transaction=s.loc[s.groupby('Month')['Amount'].idxmax()]
highest_transaction
```

```python
s["Trend"]=s.groupby("Region")["Amount"].diff()
def trendanalysis(Trend):
    if Trend>0:
        return "Increase"
    elif Trend<0:
        return "Decrease"
    elif Trend==0:
        return "Same"
    else:
        return "N/A"

s["SpendingTrend"] = s["Trend"].apply(trendanalysis)
s
```
